```python
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename = '/content/drive/MyDrive/Coding Dojo/04 Week 4: Exploratory Viz/Assignments Exercise/Suicide Deaths by Age.csv'
df = pd.read_csv(filename)
df.head()
```

```python
Year_Suicides = df.groupby(['Year'])['Suicides'].sum()
```

```python
Age_Suicides = df.groupby(['Age'])['Suicides'].sum()
```

```python
uniqueYears = list(Year_Suicides.index)
uniqueDeaths = list(Year_Suicides.values)
```

```python
uniqueAge = list(Age_Suicides.index)
uniqueAgeDeath = list(Age_Suicides.values)
```

```python
plt.style.use('seaborn')
plt.plot(uniqueYears, uniqueDeaths, c='k', marker='.', markersize=10)
plt.xlabel('Years', fontsize=15)
plt.ylabel('Number of Suicides', fontsize=15)
plt.title('Number of Suicides Per Year', fontsize=15);
print('The number of suicides gradually decreases throughout the years up until 2007.')
```

```python
sns.set_context('talk')
sns.scatterplot(x='Age', y='Suicides', data=df);
sns.lineplot(x='Age', y='Suicides', err_style='bars', color='red', data=df);
plt.title('Rate of Suicides by Age')
print('Suicide is less common on ages 20, but peaks between 30-50 years old. Afterwards, the rate of suicide begins to decline as age increases.')
```
